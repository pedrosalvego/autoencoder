```python
import pandas as pd
import os
import shutil

# 1. Configurações iniciais
csv_path = 'flood_images_annot_2.csv'
diretorio_origem = '/home/psalvego/enoe' 
diretorio_destino = '/home/psalvego/enoe_organizado' 

# 2. Carregar o CSV
df = pd.read_csv(csv_path)

# 3. Mapear todas as imagens existentes nas subpastas
print("Mapeando imagens nas subpastas... Isso pode levar alguns segundos.")
mapa_arquivos = {}

for root, dirs, files in os.walk(diretorio_origem):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
            mapa_arquivos[file] = os.path.join(root, file)

print(f"Total de {len(mapa_arquivos)} imagens mapeadas. Iniciando a organização...")

# 4. Iterar pelo CSV e mover os arquivos
fotos_encontradas = 0
fotos_nao_encontradas = 0

for index, row in df.iterrows():
    caminho_csv = row['path']
    classe = str(row['level']).strip() 
    
    # Extrai apenas o NOME do arquivo do CSV
    nome_arquivo_csv = os.path.basename(caminho_csv)
    
    # --- A MÁGICA ACONTECE AQUI ---
    # Define o caminho da pasta dessa classe específica
    pasta_da_classe = os.path.join(diretorio_destino, classe)
    
    # Cria a pasta caso ela não exista (o exist_ok=True impede erro se já existir)
    os.makedirs(pasta_da_classe, exist_ok=True)
    # ------------------------------
    
    destino_completo = os.path.join(pasta_da_classe, nome_arquivo_csv)
    
    # Verifica se o arquivo do CSV existe no nosso mapa
    if nome_arquivo_csv in mapa_arquivos:
        origem_completa = mapa_arquivos[nome_arquivo_csv]
        
        # Copia o arquivo
        shutil.copy2(origem_completa, destino_completo)
        fotos_encontradas += 1
    else:
        print(f"Aviso: Arquivo do CSV não encontrado nas subpastas: {nome_arquivo_csv}")
        fotos_nao_encontradas += 1

print("\n--- Resumo ---")
print(f"Concluído! {fotos_encontradas} fotos copiadas com sucesso.")
print(f"Falhas: {fotos_nao_encontradas} fotos não encontradas.")
```
